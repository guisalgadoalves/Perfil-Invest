```python
def high_dy(x):
    acoes_listadas = pd.read_excel("C:/arquivospec/listadas.xlsx")
    acoes_dy = acoes_listadas.sort_values (by='DY',ascending=False)
    acoes_dy = acoes_dy[['TICKER', 'DY']]
    
    print(acoes_dy.head(x))


def low_volat(x):

    empresas_listadas = pd.read_excel("C:/arquivospec/nomes_listadas_teste.xlsx")
    #display(empresas_listadas)
    empresas_listadas.insert(1, "VOLATILIDADE",'')
    for TICKER in empresas_listadas['TICKER']: 
        #print(f"{TICKER}:")
        df = web.DataReader(f'{TICKER}.SA', data_source='yahoo', start='01-01-2015', end='today')
        volat = df["Adj Close"].std()
        #print(f"Volatilidade:  {round (volat,3)}%")
        for index, row in empresas_listadas.iterrows():
            if (row['VOLATILIDADE'] == ''):
                empresas_listadas.loc[[index], ['VOLATILIDADE']] = round (volat,3)
                break
        #display(empresas_listadas)
    acoes_volat = empresas_listadas.sort_values (by='VOLATILIDADE',ascending=True)
    acoes_volat = acoes_volat[['TICKER', 'VOLATILIDADE']]
    
    print(acoes_volat.head(x))    
    
        
def high_volat(x):

    empresas_listadas = pd.read_excel("C:/arquivospec/nomes_listadas_teste.xlsx")
    #display(empresas_listadas)
    empresas_listadas.insert(1, "VOLATILIDADE",'')
    for TICKER in empresas_listadas['TICKER']: 
        #print(f"{TICKER}:")
        df = web.DataReader(f'{TICKER}.SA', data_source='yahoo', start='01-01-2015', end='today')
        volat = df["Adj Close"].std()
        #print(f"Volatilidade:  {round (volat,3)}%")
        for index, row in empresas_listadas.iterrows():
            if (row['VOLATILIDADE'] == ''):
                empresas_listadas.loc[[index], ['VOLATILIDADE']] = round (volat,3)
                break
        #display(empresas_listadas)
    acoes_volat = empresas_listadas.sort_values (by='VOLATILIDADE',ascending=False)
    acoes_volat = acoes_volat[['TICKER', 'VOLATILIDADE']]
    
    print(acoes_volat.head(x))    
    

def high_growth(x):

    empresas_listadas = pd.read_excel("C:/arquivospec/nomes_listadas_teste.xlsx")
    empresas_listadas.insert(1, "CRESCIMENTO",'')
    for TICKER in empresas_listadas['TICKER']:
        #print(f"{TICKER}:")
        df = web.DataReader(f'{TICKER}.SA', data_source='yahoo', start='01-01-2015', end='today')
        cresc = (df['Adj Close'] / df['Adj Close'].shift(1)) - 1
        #print(cresc)
        avg_growth = cresc.mean() * 250
        #print(avg_growth)
        for index, row in empresas_listadas.iterrows():
            if (row['CRESCIMENTO'] == ''):
                empresas_listadas.loc[[index], ['CRESCIMENTO']] = round (avg_growth,3)
                break
    acoes_growth = empresas_listadas.sort_values (by='CRESCIMENTO',ascending=False)
    acoes_growth = acoes_growth[['TICKER', 'CRESCIMENTO']]
    
    print(acoes_growth.head(x))


# se for 1,1,1 Conservador
# se for 1,1,2 Conservador Moderado
# se for 1,2,1 Conservador Conservador
# se for 1,2,2 Conservador1-
# se for 2.1.1 Moderado
# se for 2,1,2 Moderado Agressivo
# se for 2,2,1 Moderado Conservador
# se for 2,2,2 Moderado Conservador 
# se for 3,1,1 Agressivo 
# se for 3,1,2 Agressivo Agressivo
# se for 3,2,1 Agressivo Moderado
# se for 3,2,2 Agressivo 

```
